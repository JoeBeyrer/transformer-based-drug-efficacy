# Model Design

```python
import torch
import torch.nn as nn
import math

# PositionalEncoding
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.pe = pe.unsqueeze(1)  # (max_len, 1, d_model)
    
    def forward(self, x):
        # x: (seq_len, batch_size, d_model)
        return x + self.pe[:x.size(0)]

# MultiTaskDrugRegressor: IC50, AUC regression (separately)
class MultiTaskDrugRegressor(nn.Module):
    def __init__(self, vocab_size, d_model, nhead, num_layers, num_omics):
        super(MultiTaskDrugRegressor, self).__init__()
        # SMILES 처리 (Transformer 기반)
        self.embedding = nn.Embedding(vocab_size, d_model, padding_idx=0)
        self.positional_encoding = PositionalEncoding(d_model)
        encoder_layer = nn.TransformerEncoderLayer(d_model, nhead)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
        # Omics 데이터 처리
        self.omics_fc = nn.Linear(num_omics, d_model)
        
        # 결합된 표현은 2*d_model 차원
        # 각각의 회귀 헤드: IC50과 AUC 예측 (출력은 단일값)
        self.ic50_head = nn.Linear(2*d_model, 1)
        self.auc_head = nn.Linear(2*d_model, 1)
        
    def forward(self, smiles_tokens, omics_data):
        """
        smiles_tokens: (batch_size, seq_len)  -- 정수 인덱스로 구성 (패딩 토큰은 0)
        omics_data: (batch_size, num_omics)
        """
        # SMILES 처리: 패딩 마스크 생성
        src_key_padding_mask = (smiles_tokens == 0)  # (batch_size, seq_len)
        embedded = self.embedding(smiles_tokens).transpose(0, 1)  # (seq_len, batch_size, d_model)
        encoded = self.positional_encoding(embedded)
        # Transformer 인코더 (패딩 무시)
        encoded = self.transformer_encoder(encoded, src_key_padding_mask=src_key_padding_mask)
        
        # 평균 풀링: 패딩을 제외하고 각 배치에 대해 평균값 산출
        # encoded: (seq_len, batch_size, d_model)
        # 먼저, 패딩 위치를 0으로 채우고
        encoded = encoded.masked_fill(src_key_padding_mask.transpose(0,1).unsqueeze(-1), 0)
        sum_encoded = encoded.sum(dim=0)  # (batch_size, d_model)
        lengths = (~src_key_padding_mask).sum(dim=1, keepdim=True)  # (batch_size, 1)
        smiles_repr = sum_encoded / lengths.clamp(min=1)
        
        # Omics 데이터 처리
        omics_repr = self.omics_fc(omics_data)  # (batch_size, d_model)
        
        # 융합: SMILES 표현과 omics 표현을 concatenate
        fused = torch.cat([smiles_repr, omics_repr], dim=1)  # (batch_size, 2*d_model)
        
        # 두 개의 회귀 헤드로 예측
        ic50_pred = self.ic50_head(fused)  # (batch_size, 1)
        auc_pred = self.auc_head(fused)    # (batch_size, 1)
        
        return ic50_pred, auc_pred
```

```python


######################################
# 예제: Combined Dataset & Training Step
######################################
# 예시: 각 샘플은 omics, SMILES, 그리고 두 타겟 중 하나만 실제값을 가짐.
# NaN으로 표시된 타겟은 손실 계산 시 무시.
# dataset_sample: (smiles_tokens, omics_data, ic50_target, auc_target)
# 예를 들어, IC50 데이터셋의 경우 auc_target은 NaN, AUC 데이터셋의 경우 ic50_target은 NaN

# 샘플 배치 데이터 (batch_size=4, seq_len=10, num_omics=5)
# 임의로 생성한 정수 시퀀스 (패딩 토큰은 0)
smiles_tokens = torch.randint(1, 100, (4, 10))
# 예시로, 첫 두 샘플은 IC50 데이터셋, 나머지 두 샘플은 AUC 데이터셋으로 가정
# (IC50 데이터: ic50_target은 실제 값, auc_target은 NaN)
# (AUC 데이터: auc_target은 실제 값, ic50_target은 NaN)
omics_data = torch.randn(4, 5)
ic50_targets = torch.tensor([1.2, 0.8, float('nan'), float('nan')]).unsqueeze(1)
auc_targets  = torch.tensor([float('nan'), float('nan'), 0.45, 0.60]).unsqueeze(1)

# 모델 생성
model = MultiTaskDrugRegressor(vocab_size=100, d_model=128, nhead=4, num_layers=2, num_omics=5)

# Forward pass
ic50_pred, auc_pred = model(smiles_tokens, omics_data)

print("IC50 Predictions:", ic50_pred)
print("AUC Predictions:", auc_pred)

######################################
# 예제: 손실 계산 (유효한 타겟에 대해서만)
######################################
criterion = nn.MSELoss()

# 마스크를 통해 유효한 샘플만 선택 (NaN 체크)
ic50_mask = ~torch.isnan(ic50_targets)
auc_mask  = ~torch.isnan(auc_targets)

loss = 0.0
if ic50_mask.sum() > 0:
    loss_ic50 = criterion(ic50_pred[ic50_mask], ic50_targets[ic50_mask])
    loss += loss_ic50
if auc_mask.sum() > 0:
    loss_auc = criterion(auc_pred[auc_mask], auc_targets[auc_mask])
    loss += loss_auc

print("Total Loss:", loss.item())
```
