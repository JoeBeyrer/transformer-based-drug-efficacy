```python
import pubchempy as pcp

def get_pubchem_info(drug_name):
    compounds = pcp.get_compounds(drug_name, 'name')
    
    if compounds:
        compound = compounds[0]  # ì²« ë²ˆì§¸ ê²€ìƒ‰ ê²°ê³¼ ì„ íƒ
        return {
            "pubchem_id": compound.cid,
            "smiles": compound.canonical_smiles
        }
    else:
        #print(f"{drug_name}ì— ëŒ€í•œ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return None

drugIDs = []
drugNames = []
pubchemIDs = []
smiles = []

not_found_drugID = []
not_found_drugNames = []

for i in range(len(drug_for_AUC)):
    drugID = drug_for_AUC[i][-5:-1]
    drug_name = drug_for_AUC[i][:-13]
    info = get_pubchem_info(drug_name)

    if info:
        #print(f"{drug_name}ì˜ PubChem ID: {info['pubchem_id']}")
        #print(f"{drug_name}ì˜ SMILES: {info['smiles']}")
    
        drugIDs.append(drugID)
        drugNames.append(drug_name)
        pubchemIDs.append(info['pubchem_id'])
        smiles.append(info['smiles'])
    else:
        not_found_drugID.append(drugID)
        not_found_drugNames.append(drug_name)
        print(drugID, drug_name)

all_drugs_in_IC50_AUC = pd.DataFrame({"drugID":drugIDs, "drugName": drugNames, "pubchemID":pubchemIDs, "smiles":smiles})
```

```python

import pubchempy as pcp

def get_smiles_from_pubchem(pubchem_id):
    """PubChem IDë¡œ SMILES êµ¬ì¡°ì‹ì„ ê°€ì ¸ì˜¤ëŠ” í•¨ìˆ˜"""
    try:
        compound = pcp.Compound.from_cid(pubchem_id)
        return compound.canonical_smiles if compound else None
    except:
        return None

not_found_smiles = []
for i in range(len(not_found_pubchem)):
    pubchem_id = not_found_pubchem[i]
    smiles = get_smiles_from_pubchem(pubchem_id)
    if smiles:
        not_found_smiles.append(smiles)
    else:
        not_found_smiles.append("-")
        print(pubchem_id)


not_found_drugs_in_IC50_AUC = pd.DataFrame({"drugID":not_found_drugID, "drugName": not_found_drugNames, "pubchemID":not_found_pubchem, "smiles":not_found_smiles})
```

```python
final_all_drugs = pd.concat([all_drugs_in_IC50_AUC, not_found_drugs_in_IC50_AUC], ignore_index=True)
```
